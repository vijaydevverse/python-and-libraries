```python
from abc import ABC, abstractmethod
class Vehicle(ABC): #abstract class

  @abstractmethod
  def start_engine(self):
    pass #abstract method with no implementation

  @abstractmethod
  def stop_engine(self):
    pass #abstract method with no implementation

# making child class for abstract class instantiating

class car(Vehicle): # child class of vehicle

  def start_engine(self):  # also we need to create same function (method) in child class  other wise it will shows error!
    print("Car engine started")

  def stop_engine(self):   # also we need to create same function (method) in child class  other wise it will shows error!
    print("Car engine stopped")

car=car()
car.start_engine()
car.stop_engine()
```

 - `If we want to make object from abstract class we can't directly make object from the abstract class.`

 - `So we need to make child class for abstract class object making`
 - `And also need to make same method ( function ) on child class also`

- All child classes **must** impement the abstract methos of the parent abstract class.
- Abstract class can have both **abstract methods and concrete ( regular ) methods**.

```python
from abc import ABC, abstractmethod
class Shape(ABC):

  @abstractmethod
  def calculate_area(self):
    pass

class circle(Shape):
  def __init__(self,radius):
    self.radius = radius

  def calculate_area(self):
    print(f"circle area is: {3.14 * (self.radius)**2 }")

class Rectangle(Shape):
  def __init__(self,width,height):
     self.width = width
     self.height = height


  def calculate_area(self):
    print(f"Rectangle area is: {self.width * self.height}")

cir=circle(10)
rect=Rectangle(5,10)

cir.calculate_area()
print()
rect.calculate_area()
```

```python
# Define a class that will act as an iterator
class MyNumbers:
    # The __iter__ method is called when iteration starts
    def __iter__(self):
        self.a=1       # Initialize the starting number
        return self      # The object itself is returned as an iterator

    # The __next__ method returns the next value in the sequence
    def __next__(self):
        x=self.a       # Store the current value to return
        self.a+=1      # Increment for the next call
        return x         # Return the current value

# Create an instance of the class
myclass=MyNumbers()

# Create an iterator object from the class instance
myiter=iter(myclass)

# Retrieve and print the next value from the iterator
print(next(myiter))  # Output: 1
print(next(myiter))  # Output: 2
print(next(myiter))  # Output: 3
print(next(myiter))  # Output: 4
print(next(myiter))  # Output: 5
```

#Generator

```python
# This function is a generator function
def greeting():
    # First print statement
    print('Hi!')
    # Yield pauses the function here and returns 1 to the caller
    yield 1

    # Execution resumes here when the generator is called next()
    print("How are you?")
    # Yield pauses again and returns 2
    yield 2

    # Execution resumes here on the next call
    print("Are you there?")
    # Yield pauses and returns 3
    yield 3
```

```python
# Define the generator function
def messenger():
    # First pause: print 'Hi!' then yield 1
    print("Hi!")
    yield 1

    # Second pause: print 'How are you?' then yield 2
    print("How are you?")
    yield 2

    # Third pause: print 'Are you there?' then yield 3
    print("Are you there?")
    yield 3

# Create a generator object
messenger=messenger()

# First call to next() will execute until the first yield
result=next(messenger)  # Prints 'Hi!' and returns 1
print(result)             # Output: 1

# Second call to next() will resume from the previous yield
result=next(messenger)  # Prints 'How are you?' and returns 2
print(result)             # Output: 2

# Third call to next() will resume again
result=next(messenger)  # Prints 'Are you there?' and returns 3
print(result)             # Output: 3
```

#RegEx in Python

Example: Search the string to see if it starts with "The" and ends with "Spain":


```python
import re  # Import the 're' module, which provides support for regular expressions in Python

# Define the text to be searched
txt="The rain in Spain"

# Use the re.search() function to look for a pattern in the string
# ^The       means the string must start with "The"
# .*         means any number of any characters (0 or more)
# Spain$     means the string must end with "Spain"
# So the whole pattern "^The.*Spain$" matches any string that starts with "The" and ends with "Spain"
x=re.search("^The.*Spain$", txt)

# Check if a match was found
if x:
  # If x is not None, it means the pattern matched the string
  print("YES! We have a match!")
else:
  # If x is None, no match was found
  print("No match")
```

###The search() Function

The search() function searches the string for a match, and returns a Match object if there is a match.

If there is more than one match, only the first occurrence of the match will be returned:

Example: Search for the first white-space character in the string:

```python
import re  # Import the 're' module (used for working with regular expressions)

# Define the text where we want to search
txt="The rain in Spain"

# Use re.search() to look for the first occurrence of a whitespace character
# r"\s"  the 'r' before the string makes it a *raw string*.
#          This tells Python not to treat '\' as a special escape character.
# \s  in regular expressions, it matches any whitespace character
#       (like a space, tab, or newline).
# re.search()  searches through the string and returns a match object for the *first* match found.
x=re.search(r"\s", txt)

# The .start() method returns the position (index) in the string
# where the first whitespace character was found.
# In "The rain in Spain", the first space appears after "The", which is at index 3.
print("The first white-space character is located in position:", x.start())
```

#Multithreading

Example 1: Basic Thread

```python
import threading  # Import the threading module to work with threads

# Define a function that will be executed by the thread
def task():
    print("Thread running")  # This message will be printed when the thread runs

# Create a Thread object
# 'target=task' means that the thread will execute the 'task()' function when started
thread=threading.Thread(target=task)

# Start the thread
# This tells Python to begin executing the 'task()' function in a new thread
thread.start()

# Wait for the thread to complete
# 'join()' blocks the main program until the thread finishes running
thread.join()
```

Example 2: Multiple Threads

```python
import threading  # Import the threading module to create and manage threads

# Define the first task (function) that a thread will execute
def task1():
    print("Task 1")  # This will print when thread t1 runs

# Define the second task (function) that another thread will execute
def task2():
    print("Task 2")  # This will print when thread t2 runs

# Create a Thread object that will run the 'task1' function
t1=threading.Thread(target=task1)

# Create another Thread object that will run the 'task2' function
t2=threading.Thread(target=task2)

# Start the first thread  it begins executing 'task1()' in parallel
t1.start()

# Start the second thread  it begins executing 'task2()' in parallel
t2.start()

# Wait for the first thread to finish before continuing
t1.join()

# Wait for the second thread to finish before continuing
t2.join()
```

Python program to download multiple files concurrently using threads

```python
import threading   # Import the threading module to create and manage threads
import time        # Import time module to simulate delay (like downloading time)

# Define a function that simulates downloading a file
def download_file(file):
    print(f"Downloading {file}...")  # Indicate which file is being downloaded
    time.sleep(2)                    # Simulate a 2-second download time
    print(f"{file} downloaded.")     # Indicate the file has finished downloading

# List of files to "download"
files=["file1.txt", "file2.txt", "file3.txt"]

# Create a list of Thread objects  one thread per file
# Each thread runs the 'download_file' function with a file name as its argument
threads = [threading.Thread(target=download_file, args=(f,)) for f in files]

# Start all threads (downloads start concurrently)
for t in threads:
    t.start()

# Wait for all threads to finish before ending the program
for t in threads:
    t.join()
```

Downloading chunk 4...
Downloading chunk 5...
Download complete. Starting virus scan...
Scanning file for viruses...
Scanning...

```python
import threading
import time

def download_file(event):
    print("File download started...")
    for i in range(5):
        print(f"Downloading chunk {i+1}...")
        time.sleep(1)
    print("Download complete. Starting virus scan...")
    event.set()

def virus_scan(event):
    event.wait()
    print("Scanning file for viruses...")
    for i in range(3):
        print(f"Scanning... {i+1} seconds")
        time.sleep(1)
    print("Scan complete. File is safe to use.")

    print("Download process initiated...")
    time.sleep(1)

download_event=threading.Event()

download_thread=threading.Thread(target=download_file, args=(download_event,))
scan_thread=threading.Thread(target=virus_scan, args=(download_event,))

download_thread.start()
scan_thread.start()
download_thread.join()
scan_thread.join()

print("All download and scanning operations completed successfully.")
```

```python
import threading
import time

def cook_rice(event):
    print("Cooking rice started...")
    for i in range(5):
        if i==1:
            print("Starting to cook curry...")
            curry_thread=threading.Thread(target=cook_curry, args=(event,))
            curry_thread.start()
        print(f"Cooking rice... {i+1} seconds")
        time.sleep(1)
    print("Rice cooking finished.")
    event.set()

def cook_curry(event):
    print("Cooking curry started...")
    for i in range(4):
        print(f"Cooking curry... {i+1} seconds")
        time.sleep(1)
    print("Curry cooking finished.")
    event.set()

    print("Meal preparation started...")
    time.sleep(1)

rice_done=threading.Event()
curry_done=threading.Event()

rice_thread=threading.Thread(target=cook_rice, args=(curry_done,))
rice_thread.start()

rice_thread.join()
curry_done.wait()

print("Both rice and curry are ready.")
print("Meal is served!")
```

##Raise Exception

The raise statement in python is used to manually trigger an exception when a specific error condition occurs.

```python
# Ask the user to enter their age and convert the input to an integer
# Note: If the user types something that is not a number (e.g., "abc"),
# Python will immediately raise a ValueError before reaching the next lines.
age = int(input("Enter your age: "))

# Check if the entered age is negative
if age < 0:
    # If the number is negative, we raise a ValueError ourselves.
    # This stops the program immediately and shows the custom error message.
    raise ValueError("Age cannot be negative")
else:
    # If the age is valid (0 or positive), this line prints it.
    print(f"Your age is {age}")

# This line will only run if no ValueError occurred above.
# If an error happens (either from invalid input or from our 'raise'),
# this line will NOT execute.
print("No try except")
```

```python
try:
    # Ask the user to enter their age and convert it to an integer
    # If the user types something that is not a number (e.g., "abc"),
    # Python automatically raises a ValueError here.
    age = int(input("Enter your age: "))

    # Check if the entered age is negative
    if age < 0:
        # Manually raise a ValueError with a custom message
        # if the user enters a negative number.
        raise ValueError("Age cannot be negative")
    else:
        # If no error occurs and the age is valid, print it
        print(f"Your age is {age}")

# The 'except' block runs only if a ValueError happens in the try block.
# The variable 'e' stores the error message (either from int() or from our raise statement).
except ValueError as e:
    # Print the error message instead of crashing the program
    print(e)

# This line runs no matter what  even if an error happened.
# It shows that the program continues after handling the error.
print("try except finished")
```

## Custom Exceptions




Custom exceptions are user- defined exceptions in Python that help handle domain-specific errors more meaningully and improve code clarity.

**Example Code :**

    class CustomError(Exception)"
        pass
    raise CustomError("This is a custom error")

```python
# Define a custom exception class
# 'InvalidAgeException' inherits from Python's built-in 'Exception' class.
# This allows us to create our own specific type of error.
class InvalidAgeException(Exception):
    "Raised when the input value is less than 18"
    pass  # 'pass' means no extra code inside the class  it just defines the exception type.

# The number we are comparing to (minimum age required to vote)
number = 18

try:
    # Ask the user to enter a number and convert it to an integer.
    input_num = int(input("Enter a number: "))

    # Check if the entered number (age) is less than 18.
    if input_num < number:
        # If condition is true, manually raise our custom exception.
        raise InvalidAgeException
    else:
        # If age is 18 or older, print eligibility message.
        print("Eligible to vote")

# This block runs if an InvalidAgeException is raised above.
except InvalidAgeException:
    # Handle the error gracefully instead of crashing the program.
    print("Exception Occurred: Invalid age")
```

* Uses tryexcept to catch ZeroDivisionError

* Prints "Cannot divide by zero." if it occurs

* Otherwise, prints the result using the else block

* Prints "Operation complete." in a finally block

```python
def safe_divide(a,b):
    try:
        result=a/b
    except ZeroDivisionError:
        print("Cannot divide by zero.")
    else:
        print("Result:", result)
    finally:
        print("Operation complete.")
x=int(input("Enter the numerator: "))
y=int(input("Enter the denominator: "))
safe_divide(x,y)
```

##3. File Handling Error

Goal: Handle missing files.
Task:
* Write a function that tries to open a file for reading.
* If the file does not exist, catch the FileNotFoundError and print "File not found!".
* Always close the file (or print "Closing file" in finally) even if an error occurs.

```python
def read_file(filename):
    file=None
    try:
        file=open(filename, 'r')
        content=file.read()
        print("File content:\n", content)
    except FileNotFoundError:
        print("File not found!")
    finally:
        if file:
            file.close()
            print("Closing file.")

filename=input("Enter the filename to read: ")
read_file(filename)
```
